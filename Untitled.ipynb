```python
import matplotlib.pyplot as plt
```

```python
savings = [0, 50, 100, 150, 200, 100, 500, 200, 350, 600, 750, 1000]

months = [ 'jan', 'feb', 'mar', 'apr', 'may', 'jun', 'jul', 'aug', 'sep', 'oct', 'nov', 'dec']
```

```python
plt.plot(months, savings)
plt.xlabel('Month')
plt.ylabel('Savings $')
plt.title("Savings Account")
plt.show()
```
